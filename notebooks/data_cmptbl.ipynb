### Just a cleaning fct

```python
import re
from unidecode import unidecode
import ast

def clean_artist_list(artist_str):
    """Extract artist name from list string"""
    if pd.isna(artist_str):
        return ""
    
    try:
        artist_list = ast.literal_eval(artist_str)
        artist = artist_list[0] if artist_list else ""
        return clean_text(artist)
    except:
        return clean_text(str(artist_str))

def clean_text(text):
    """Normalize text for better matching"""
    if pd.isna(text):
        return ""
    
    text = str(text).lower()
    text = unidecode(text)  # Remove accents
    text = re.sub(r'\(.*?\)', '', text)  # Remove parentheses
    text = re.sub(r'\[.*?\]', '', text)  # Remove brackets
    text = re.sub(r' - .*', '', text)  # Remove dash suffixes
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = re.sub(r'\s+', ' ', text)  # Normalize spaces
    text = text.strip()
    text = re.sub(r'^(the|a|an)\s+', '', text)  # Remove articles
    
    return text

```

```python
# Apply cleaning
sp_df['name_clean'] = sp_df['name'].apply(clean_text)
sp_df['artist_clean'] = sp_df['artists'].apply(clean_artist_list)  # Special handling!

fm_df['title_clean'] = fm_df['track_title'].apply(clean_text)
fm_df['artist_clean'] = fm_df['artist_name'].apply(clean_text)

# For fm_df_cleaned - fix the warning AND clean
fm_df_cleaned = fm_df[fm_df["track_id"].isin(id_music_list_df["track_id"])].copy()
fm_df_cleaned['title_clean'] = fm_df_cleaned['track_title'].apply(clean_text)
fm_df_cleaned['artist_clean'] = fm_df_cleaned['artist_name'].apply(clean_text)
```

```python
df_matched = pd.merge(
    sp_df, 
    fm_df_cleaned,  # Use cleaned (8K tracks with audio)
    left_on=['name_clean', 'artist_clean'],
    right_on=['title_clean', 'artist_clean'],
    how='inner'
)

print(f"Spotify dataset: {len(sp_df):,} tracks")
print(f"FMA Small (with audio): {len(fm_df_cleaned):,} tracks")
print(f"Matches found: {len(df_matched):,}")
```
