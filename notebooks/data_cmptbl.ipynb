### Just a cleaning fct

```python
import re
from unidecode import unidecode
import ast

def clean_artist_list(artist_str):
    """Extract artist name from list string"""
    if pd.isna(artist_str):
        return ""
    
    try:
        artist_list = ast.literal_eval(artist_str)
        artist = artist_list[0] if artist_list else ""
        return clean_text(artist)
    except:
        return clean_text(str(artist_str))

def clean_text(text):
    """Normalize text for better matching"""
    if pd.isna(text):
        return ""
    
    text = str(text).lower()
    text = unidecode(text)  # Remove accents
    text = re.sub(r'\(.*?\)', '', text)  # Remove parentheses
    text = re.sub(r'\[.*?\]', '', text)  # Remove brackets
    text = re.sub(r' - .*', '', text)  # Remove dash suffixes
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = re.sub(r'\s+', ' ', text)  # Normalize spaces
    text = text.strip()
    text = re.sub(r'^(the|a|an)\s+', '', text)  # Remove articles
    
    return text

```

```python
# Apply cleaning
sp_df['name_clean'] = sp_df['name'].apply(clean_text)
sp_df['artist_clean'] = sp_df['artists'].apply(clean_artist_list)  # Special handling!

fm_df['title_clean'] = fm_df['track_title'].apply(clean_text)
fm_df['artist_clean'] = fm_df['artist_name'].apply(clean_text)

# For fm_df_cleaned - fix the warning AND clean
fm_df_cleaned = fm_df[fm_df["track_id"].isin(id_music_list_df["track_id"])].copy()
fm_df_cleaned['title_clean'] = fm_df_cleaned['track_title'].apply(clean_text)
fm_df_cleaned['artist_clean'] = fm_df_cleaned['artist_name'].apply(clean_text)
```

```python
from tqdm import tqdm
import shutil
from loguru import logger
import re
from unidecode import unidecode
import ast

def clean_artist_list(artist_str):
    """Extract artist name from list string"""
    if pd.isna(artist_str):
        return ""
    try:
        artist_list = ast.literal_eval(artist_str)
        artist = artist_list[0] if artist_list else ""
        return clean_text(artist)
    except:
        return clean_text(str(artist_str))

def clean_text(text):
    """Normalize text for better matching"""
    if pd.isna(text):
        return ""
    text = str(text).lower()
    text = unidecode(text)
    text = re.sub(r'\(.*?\)', '', text)
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r' - .*', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text)
    text = text.strip()
    text = re.sub(r'^(the|a|an)\s+', '', text)
    return text

def extract_matched_subset():
    FMA_LARGE_DIR = cfg.paths.fma_large
    FMA_METADATA = cfg.paths.fma_metadata
    SPOTIFY_CSV = cfg.paths.spotify / "tracks_features.csv"

    OUTPUT_DIR = cfg.paths.data / "processed"
    OUTPUT_AUDIO_DIR = OUTPUT_DIR / "audio"
    OUTPUT_METADATA  = OUTPUT_DIR / "matched_metadata.csv"

    OUTPUT_AUDIO_DIR.mkdir(parents=True, exist_ok=True)

    fma_df = pd.read_csv(FMA_METADATA / "raw_tracks.csv")
    fma_df = fma_df[["track_id", "album_title", "artist_name", "track_title"]]
    
    sp_df = pd.read_csv(SPOTIFY_CSV)
    sp_df = sp_df.drop(columns=["id", "album_id", "artist_ids"], errors='ignore')

    logger.info(f"    FMA metadata: {len(fma_df):,} tracks")
    logger.info(f"    Spotify data: {len(sp_df):,} tracks")

    sp_df['name_clean'] = sp_df['name'].apply(clean_text)
    sp_df['artist_clean'] = sp_df['artists'].apply(clean_artist_list)
    
    fma_df['title_clean'] = fma_df['track_title'].apply(clean_text)
    fma_df['artist_clean'] = fma_df['artist_name'].apply(clean_text)

    logger.info("    Text cleaned and normalized")

    audio_files = []
    for root, dirs, files in os.walk(FMA_LARGE_DIR):
        for file in files:
            if file.endswith('.mp3'):
                audio_files.append(int(Path(file).stem))

    logger.info(f"    Found {len(audio_files):,} audio files in FMA Large")
    
    fma_with_audio = fma_df[fma_df['track_id'].isin(audio_files)].copy()

    logger.info(f"    {len(fma_with_audio):,} FMA tracks have audio files")

    matched_df = pd.merge(
        sp_df,
        fma_with_audio,
        left_on=['name_clean', 'artist_clean'],
        right_on=['title_clean', 'artist_clean'],
        how='inner',
        suffixes=('_spotify', '_fma')
    )

    logger.info(f"    Found {len(matched_df)} FMA-Spotify matches!")

    existing_tracks = []
    for idx, row in matched_df.iterrows():
        track_id = row['track_id']
        folder = str(track_id).zfill(6)[:3]
        source = FMA_LARGE_DIR / folder / f"{track_id}.mp3"
        
        if source.exists():
            existing_tracks.append(idx)
    
    # Keep only tracks with existing files
    matched_df = matched_df.loc[existing_tracks].reset_index(drop=True)
    logger.info(f"    {len(matched_df)} tracks have audio files available")
    
    matched_df = matched_df.drop(columns=['name_clean', 'title_clean'], errors='ignore')    

    id_cols = ['track_id']
    fma_cols = ['track_title', 'artist_name', 'album_title']
    spotify_cols = ['name', 'artists', 'album', 'year', 'release_date',
                    'danceability', 'energy', 'key', 'loudness', 'mode',
                    'speechiness', 'acousticness', 'instrumentalness',
                    'liveness', 'valence', 'tempo', 'duration_ms', 'time_signature']
    
    col_order = id_cols + fma_cols + [col for col in spotify_cols if col in matched_df.columns]
    col_order = [col for col in col_order if col in matched_df.columns]
    matched_df = matched_df[col_order]

    copied_count = 0
    failed_copies = 0
    logger.info(f"\n[4/4] Copying {len(matched_df)} audio files...")
    logger.info(f"   From: {FMA_LARGE_DIR}")
    logger.info(f"   To:   {OUTPUT_AUDIO_DIR}")
    
    for track_id in tqdm(matched_df["track_id"], desc="     Coping files"):
        folder = str(track_id).zfill(6)[:3]
        source = FMA_LARGE_DIR / folder / f"{track_id}.mp3"

        if source.exists():
            dest = OUTPUT_AUDIO_DIR / f"{track_id}.mp3"
            shutil.copy2(source, dest)
            copied_count += 1
        else:
            logger.info(f"    Missing: {track_id}.mp3")
            failed_copies += 1
    
    matched_df.to_csv(OUTPUT_METADATA, index=False)
    logger.info(f"    Saved to: {OUTPUT_METADATA}")
    return matched_df

```
